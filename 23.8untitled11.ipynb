```python
import numpy as np

# Giả lập 5 xe taxi và 1 khách hàng
taxis = np.random.rand(5, 2) * 10
customer = np.array([5, 5])

# Heuristic: Khoảng cách Euclidean gần nhất
distances = np.linalg.norm(taxis - customer, axis=1)
best_taxi_idx = np.argmin(distances)

print(f"Khoảng cách các xe: {distances}")
print(f"Điều xe số {best_taxi_idx} đến đón khách vì khoảng cách gần nhất: {distances[best_taxi_idx]:.2f}")
```
