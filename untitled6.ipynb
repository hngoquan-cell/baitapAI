```python
from shapely.geometry import Point
import geopandas as gpd

center = Point(106.6958, 10.7828)
# Tạo các buffer bán kính tương ứng (ước tính độ kinh vĩ)
dist_3km = center.buffer(0.027)
dist_5km = center.buffer(0.045)

gdf = gpd.GeoDataFrame(geometry=[dist_5km, dist_3km], crs="EPSG:4326")
gdf.plot(color=['yellow', 'orange'], alpha=0.5, edgecolor='red')
```
