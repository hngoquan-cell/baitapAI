```python
from geopy.geocoders import Nominatim
import folium
import time

geolocator = Nominatim(user_agent="ueh_exercise")
addresses = ["17 Phạm Ngọc Thạch, Quận 3, HCM", "279 Nguyễn Tri Phương, Quận 10, HCM"]
m = folium.Map(location=[10.77, 106.68], zoom_start=13)

for addr in addresses:
    loc = geolocator.geocode(addr)
    if loc:
        folium.Marker([loc.latitude, loc.longitude], popup=addr).add_to(m)
        print(f"Địa chỉ: {addr} -> Tọa độ: {loc.latitude}, {loc.longitude}")
    time.sleep(1) # Tránh bị chặn request
m
```
