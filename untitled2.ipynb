```python
import folium

# Tọa độ UEH (Cơ sở A)
ueh_location = [10.7828, 106.6958]
m = folium.Map(location=ueh_location, zoom_start=16)

# Danh sách địa điểm lân cận
locations = [
    {"name": "UEH - Cơ sở A", "coord": [10.7828, 106.6958], "type": "Trường học"},
    {"name": "Diamond Plaza", "coord": [10.7811, 106.6989], "type": "TT Thương mại"},
    {"name": "Công viên Lê Văn Tám", "coord": [10.7841, 106.6912], "type": "Công viên"},
    {"name": "Bệnh viện Quận 1", "coord": [10.7850, 106.6940], "type": "Bệnh viện"},
    {"name": "Bưu điện Thành phố", "coord": [10.7799, 106.6999], "type": "Cơ quan hành chính"}
]

for loc in locations:
    folium.Marker(
        location=loc["coord"],
        popup=f"{loc['name']} ({loc['type']})",
        tooltip="Click để xem"
    ).add_to(m)

m
```
