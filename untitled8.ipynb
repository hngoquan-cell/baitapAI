```python
import networkx as nx

# Sử dụng đồ thị từ bài 23.6
G = nx.grid_2d_graph(6, 6)
source = (0, 0)
target = (5, 5)

# Thuật toán Dijkstra
path_dijkstra = nx.dijkstra_path(G, source, target)

# Thuật toán A* (với hàm heuristic là khoảng cách Manhattan)
def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])
path_astar = nx.astar_path(G, source, target, heuristic=heuristic)

print(f"Đường đi Dijkstra: {path_dijkstra}")
print(f"Đường đi A*: {path_astar}")
print("Nhận xét: Cả hai thuật toán đều tìm ra lộ trình tối ưu, nhưng A* hiệu quả hơn về mặt tính toán.")
```
