```python
def greedy_delivery(start_node, locations):
    route = [start_node]
    current = start_node
    remaining = locations.copy()

    while remaining:
        # Tìm điểm gần nhất tiếp theo
        next_node = min(remaining, key=lambda x: np.linalg.norm(np.array(current) - np.array(x)))
        route.append(next_node)
        remaining.remove(next_node)
        current = next_node
    return route

points = [(2,3), (8,1), (5,9), (1,7)]
print(f"Lộ trình giao hàng tối ưu (Greedy): {greedy_delivery((0,0), points)}")
```
