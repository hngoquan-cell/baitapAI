```python
def analyze_traffic_risk(density, weather_impact):
    # Logic mờ đơn giản: Mật độ (0-1), Ảnh hưởng thời tiết (0-1)
    risk_score = (density * 0.7) + (weather_impact * 0.3)

    if risk_score > 0.7: return "Nguy cơ Cao - Đề xuất tuyến đường thay thế"
    elif risk_score > 0.4: return "Nguy cơ Trung bình"
    return "Nguy cơ Thấp"

print(f"Trạng thái khu vực A: {analyze_traffic_risk(0.85, 0.6)}")
```
