```python
import networkx as nx
import matplotlib.pyplot as plt

# Tạo một mạng lưới giao thông giả lập (Grid graph) đại diện cho khu đô thị
G = nx.grid_2d_graph(5, 5)
pos = {node: node for node in G.nodes()} # Tọa độ các nút

plt.figure(figsize=(8, 6))
nx.draw(G, pos, node_size=50, node_color='blue', edge_color='gray', with_labels=False)
plt.title("Mô phỏng mạng lưới giao thông đô thị thông minh")
plt.show()

print(f"Số lượng nút giao: {G.number_of_nodes()}")
print(f"Số lượng đoạn đường: {G.number_of_edges()}")
```
