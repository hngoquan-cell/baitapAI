```python
import folium
from folium.plugins import HeatMap
import numpy as np

m = folium.Map(location=[10.77, 106.68], zoom_start=13)
# Giả lập 100 điểm giao dịch
data = np.random.normal(size=(100, 2), loc=[10.77, 106.68], scale=0.02).tolist()

HeatMap(data).add_to(m)
m
```
