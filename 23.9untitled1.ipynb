```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Giả lập 100 điểm nhu cầu khách hàng
X = np.random.rand(100, 2) * 100

# Phân thành 4 cụm để tìm 4 vị trí đặt trạm kho hàng/xe
kmeans = KMeans(n_clusters=4, n_init=10).fit(X)
centers = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Trạm tối ưu')
plt.title("Tối ưu hóa vị trí đặt kho/trạm bằng K-Means")
plt.legend()
plt.show()
```
