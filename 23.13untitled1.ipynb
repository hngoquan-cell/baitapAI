```python
import folium

# Tạo dashboard bản đồ với các lớp dữ liệu khác nhau
m = folium.Map(location=[10.7626, 106.6602], zoom_start=13)

# Lớp 1: Điểm giao dịch
folium.Marker([10.7626, 106.6602], popup="Trụ sở chính").add_to(m)

# Lớp 2: Vùng nguy cơ (Circle)
folium.Circle([10.77, 106.67], radius=1000, color='red', fill=True, popup="Vùng tắc nghẽn").add_to(m)

m
```
