```python
from sklearn.linear_model import LinearRegression
import numpy as np

# X: Giờ trong ngày (6h-10h), y: Số lượng đơn gọi xe
X = np.array([[6], [7], [8], [9], [10]])
y = np.array([15, 45, 100, 80, 60])

model = LinearRegression().fit(X, y)
prediction = model.predict([[11]]) # Dự báo nhu cầu lúc 11h

print(f"Dự báo nhu cầu lúc 11h sáng: {int(prediction[0])} đơn hàng.")
```
